```python
x1 = data['Size'].values
x2 = data['BHK'].values
x3 = data['Bathroom'].values
x4 = data['Furnishing Status'].values
y = data['Rent'].values

max_x1 = max(x1)
min_x1 = min(x1)
mean_x1 = round(np.mean(x1))

Y = [y]
Y = np.transpose(Y)
print(Y)
```

```python
def replace_values(list_to_replace, item_to_replace, item_to_replace_with):
    return [item_to_replace_with if item == item_to_replace else item for item in list_to_replace]
x4 = replace_values(x4,'Furnished',3)
x4 = replace_values(x4,'Semi-Furnished',2)
x4 = replace_values(x4,'Unfurnished',1)
```

```python
X = []
column = 5
row = len(Y)
for r in range(row):
  temp_list = []
  temp_list.append(1)
  temp_list.append(x1[r])
  temp_list.append(x2[r])
  temp_list.append(x3[r])
  temp_list.append(x4[r])
  X.append(temp_list)

print((X))
```

# Task 1

## Gradient Descent

```python
m = len(x1)
epoch = 1000
learning_rate = 0.0000001
theta_0 = 0
theta_1 = 0
theta_2 = 0
theta_3 = 0
theta_4 = 0

for i in range(epoch):
  temp_0 = 0
  temp_1 = 0
  temp_2 = 0
  temp_3 = 0
  temp_4 = 0

  for j in range(m):
    temp_0 += (theta_0 + theta_1*x1[j] + theta_2*x2[j] + theta_3*x3[j] + theta_4*x4[j] - y[j])
    temp_1 += (theta_0 + theta_1*x1[j] + theta_2*x2[j] + theta_3*x3[j] + theta_4*x4[j] - y[j]) * x1[j]
    temp_2 += (theta_0 + theta_1*x1[j] + theta_2*x2[j] + theta_3*x3[j] + theta_4*x4[j] - y[j]) * x2[j]
    temp_3 += (theta_0 + theta_1*x1[j] + theta_2*x2[j] + theta_3*x3[j] + theta_4*x4[j] - y[j]) * x3[j]
    temp_4 += (theta_0 + theta_1*x1[j] + theta_2*x2[j] + theta_3*x3[j] + theta_4*x4[j] - y[j]) * x4[j]
  
  theta_0 = theta_0 - learning_rate * (temp_0) / m
  theta_1 = theta_1 - learning_rate * (temp_1) / m
  theta_2 = theta_2 - learning_rate * (temp_2) / m
  theta_3 = theta_3 - learning_rate * (temp_3) / m
  theta_4 = theta_4 - learning_rate * (temp_4) / m

print(theta_0,theta_1,theta_2,theta_3,theta_4)
```

## Cost Calcutaion

```python
sum = 0
for j in range(m):
    theta_sum = theta_0 +  theta_1*x1[j] + theta_2*x2[j] + theta_3*x3[j] + theta_4*x4[j] - y[j]
    theta_sum = theta_sum * theta_sum
    sum = sum + theta_sum
    
cost = sum/(2*m)
print(cost)
```

## Plotting

```python
plt.rcParams['figure.figsize'] = (10.0, 5.0)

max_x = np.max(x1) + 100
min_x = np.min(x1) - 100

x = np.linspace(min_x, max_x, 1000)
yy = theta_1 * x + theta_0

plt.plot(x, yy, color='#58b970', label='Regression Line')
plt.scatter(x1, y, c = "#ef5423", label = 'Scatter Plot')

plt.xlabel('House Size')
plt.ylabel('Price')
plt.legend()
plt.show()
```

## Normal Equation

```python
XT = np.transpose(X)
XTX = np.matmul(XT,X)

from numpy.linalg import inv
INV_XTX = inv(XTX)

XTY = np.matmul(XT,Y)

thetas = np.matmul(INV_XTX,XTY)
print(len(thetas),len(thetas[0]))
print(thetas)
```

```python
t0 = thetas[0][0]
t1 = thetas[1][0]
t2 = thetas[2][0]
t3 = thetas[3][0]
t4 = thetas[4][0]
print(t0,t1,t2,t3,t4)
```

## Cost Calculation

```python
m = len(x1)
sum = 0
for j in range(m):
    tsum = t0 +  t1*x1[j] + t2*x2[j] + t3*x3[j] + t4*x4[j] - y[j]
    tsum = tsum * tsum
    sum = sum + tsum
    
cost = sum/(2*m)

prediction = t0 + t1*1100 + t2*2 + t3*2 + t4*1

print('cost: ', cost)
print('prediction: ', prediction)
```
